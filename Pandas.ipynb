```python
#Find and print the minimum, maximum and average value of the feature for each kind of flower ?
#Print the value with two decimal places. ote: Order for flower is Iris-setosa, Iris-versicolor and Iris-virginica.
#Output Format
#minSL minSW minPL minPW Iris-setosa
#maxSL maxSW maxPL maxPW Iris-setosa
#avgSL avgSW avgPL avgPW Iris-setosa
#. . .
import pandas as pd 
iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=['SepalLength','SepalWidth' ,'PetalLength', 'PetalWidth', 'Species'])
df = iris.copy()
ans_is = df[(df.Species == "Iris-setosa")].describe().loc[['min','max','mean']]
ans_ivs = df[(df.Species== "Iris-versicolor")].describe().loc[['min','max','mean']]
ans_iv = df[(df.Species == "Iris-virginica")].describe().loc[['min','max','mean']]
for i in (ans_is.values):
    for j in i :
        print(format(j,'.2f'),end=' ')
    print('Iris-setosa')
    
for i in (ans_ivs.values):
    for j in i :
        print(format(j,'.2f'),end=' ')
    print('Iris-versicolor')
    
for i in (ans_iv.values):
    for j in i :
        print(format(j,'.2f'),end=' ')
    print('Iris-virginica')
```

```python
#Find the frequency of the Casualty in Red Corridor states and in Jammu and Kashmir ?Here Frequency is (Total Casualty/Total Number of a year)
#Print frequency as integer value.
#Note:Red Corridor state includes Jharkhand, Odisha, Andhra Pradesh, and Chhattisgarh. Here Casualty=Killed +Wounded.Don't fill the nan value present in Killed and Wounded feature.
#Output Format :
#Frequency1 Frequency2 jj


import pandas as pd
import numpy as np

data = pd.read_csv('C:\\Users\\Chiranth SB\\Downloads\\terrorismData.csv')
df = data.copy()
df['cas'] = df['Killed'] + df['Wounded']
df_jk = df[(df.Country == 'India') & (df.State == 'Jammu and Kashmir')]
df_red = df[(df.Country == 'India') & (df.State == 'Jharkhand' ) | (df.State == 'Odisha') | (df.State == 'Andhra Pradesh') | (df.State == 'Chhattisgarh')]
cas_jk = df_jk.cas.sum()
#we need all the count of years passed not the count of entries in the column years
year = len(set(df.Year))  # set gives unique fig of that column
freq_jk = int(cas_jk // year)
cas_red = df_red.cas.sum()
freq_red = int(cas_red // year)
print(freq_red,freq_jk)
```
